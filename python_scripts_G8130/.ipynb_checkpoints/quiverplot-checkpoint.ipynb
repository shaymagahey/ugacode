```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import xarray as xr
import cartopy.crs as ccrs
import netCDF4 as nc
import pandas as pd
from matplotlib.pyplot import figure
```

```python
filePI = "/glade/work/smagahey/GEOG8130/data/project_results/my_cesm.B1850.f19_g17.project_preindustrial.new.cam.U.V.lev_p.0001-0010.nc"
fileCH4 = "/glade/work/smagahey/GEOG8130/data/project_results/my_cesm.B1850.f19_g17.ch4_perturbation.cam.new.h0.U.V.lev_p.0001-0010.nc"

```

```python
dsPI = xr.open_dataset(filePI)
dsCH4 = xr.open_dataset(fileCH4)
U_PI = dsPI["U"]
V_PI = dsPI["V"]
U_CH4 = dsCH4["U"]
V_CH4 = dsCH4["V"]
```

```python
upi = np.nanmean(U_PI[0:108,3,:,:],axis=0)
vpi = np.nanmean(V_PI[0:108,3,:,:],axis=0)
```

```python
upi.shape
```

```python
changeU = np.nanmean(U_CH4[:,3,:,:],axis=0) - upi
changeV = np.nanmean(V_CH4[:,3,:,:],axis=0) - vpi
```

```python
changeU
```

```python
lat = np.linspace(-90,90,96)
lon = np.linspace(0,360,144)
```

```python
LON,LAT=np.meshgrid(lon,lat)
```

```python
fig = plt.figure(figsize=(10,5))
n=3
ax1 = fig.add_subplot(1,2,1,projection=ccrs.Robinson(central_longitude=0, globe=None))
ax1.coastlines()
q = ax1.quiver(LON[::n,::n],LAT[::n,::n],upi[::n,::n],vpi[::n,::n],color="b",transform=ccrs.PlateCarree(),scale=300,linewidth=8,headlength=2,headaxislength=2,headwidth=4)
#Q = ax.quiver(LON[::n,::n],LAT[::n,::n],changeU[::n,::n],changeV[::n,::n],color="k",transform=ccrs.PlateCarree(),scale=80,linewidth=8,headlength=2,headaxislength=2,headwidth=4)
# [::n,::n] to skip n numbers of elements
#fig.colorbar(Q)
ax1.quiverkey(q,.97,.03,10,label="10 m/s",fontproperties={'size':7})
plt.title('Background 850 mb Winds') 
plt.tight_layout()
                      
ax2 = fig.add_subplot(1,2,2,projection=ccrs.Robinson(central_longitude=0, globe=None))
ax2.coastlines()
# = ax2.quiver(LON[::n,::n],LAT[::n,::n],upi[::n,::n],vpi[::n,::n],color="b",transform=ccrs.PlateCarree(),scale=300,linewidth=8,headlength=2,headaxislength=2,headwidth=4)
Q = ax2.quiver(LON[::n,::n],LAT[::n,::n],changeU[::n,::n],changeV[::n,::n],color="k",transform=ccrs.PlateCarree(),scale=80,linewidth=8,headlength=2,headaxislength=2,headwidth=4)
# [::n,::n] to skip n numbers of elements
#fig.colorbar(Q)
ax2.quiverkey(Q,.97,.03,3,label="3 m/s",fontproperties={'size':7})
plt.title('Change in 850 mb Winds') 
plt.tight_layout()
plt.savefig("/glade/work/smagahey/GEOG8130/data/project_results/quiversubplot.png")
                      
                      
```
